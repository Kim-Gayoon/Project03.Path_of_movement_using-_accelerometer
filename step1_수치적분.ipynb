```python
def df_rect(data):    # 수치적분, 직사각형 적분 가속도 -> 속도 -> 거리
  result=[0]

  for i in range(len(data)-1):
    n=result[i]+(data[i+1]*(df_t[i+1]-df_t[i]))
    result.append(round(n, 7))
  return result 
```

```python
def df_trapz(data):  # 수치적분, 사다리꼴적분 가속도 -> 속도 -> 거리
    result = [0]

    for i in range(len(data) - 1):
        n = result[i] + ((data[i + 1] + data[i]) * (df_t[i + 1] - df_t[i]) / 2)  # 수치적분, 사다리꼴적분
        result.append(round(n, 7))
    return result
```

```python
def df_simps(data):
  result = [0]

  for i in range(len(data)-1):
    rect=result[i]+ (data[i+1]*(df_t[i+1]-df_t[i]))    # 수치적분, 직사각형 적분 가속도 -> 속도 -> 거리
    trapz=result[i] + ((data[i + 1] + data[i]) * (df_t[i + 1] - df_t[i]) / 2)    # 수치적분, 사다리꼴적분 가속도 -> 속도 -> 거리
  
    n = rect*2/3 + trapz*1/3
    result.append(round(n,7))
  return result
```

```python
# 직사각형
r_x_2 = df_rect(list_x)  # x 가속도 -> 속도
r_x_3 = df_rect(r_x_2)     # x 속도 -> 거리

r_y_2 = df_rect(list_y) # y 가속도 - > 속도
r_y_3 = df_rect(r_y_2)# y 속도 -> 거리
```

```python
# 사다리꼴
t_x_2 = df_trapz(list_x)   # x 가속도 -> 속도
t_x_3 = df_trapz(t_x_2)      # x 속도 -> 거리

t_y_2 = df_trapz(list_y)   # y 가속도 - > 속도
t_y_3 = df_trapz(t_y_2)      # y 속도 -> 거리
```

```python
# 심스 
s_x_2 = df_simps(list_x)  # x 가속도 -> 속도
s_x_3 = df_simps(s_x_2)     # x 속도 -> 거리

s_y_2 = df_simps(list_y) # y 가속도 - > 속도
s_y_3 = df_simps(s_y_2)# y 속도 -> 거리
```

```python
# 직사각형 
r_df_scatter = pd.DataFrame(zip(r_x_3, r_y_3), columns=['r_m_x', 'r_m_y']) # 리스트 열단위 데이터프레임화

# 사다리꼴
t_df_scatter = pd.DataFrame(zip(t_x_3, t_y_3), columns=['t_m_x', 't_m_y']) # 리스트 열단위 데이터프레임화

# 심스
s_df_scatter = pd.DataFrame(zip(s_x_3, s_y_3), columns=['s_m_x', 's_m_y']) # 리스트 열단위 데이터프레임화
```

```python
# 직사각형
plt.scatter(r_df_scatter['r_m_x'], r_df_scatter['r_m_y'])
plt.show() # 플로팅
```

```python
# 사다리꼴
plt.scatter(t_df_scatter['t_m_x'], t_df_scatter['t_m_y'])
plt.show() # 플로팅
```

```python
# 심스 
plt.scatter(s_df_scatter['s_m_x'], s_df_scatter['s_m_y'])
plt.show() # 플로팅
```
