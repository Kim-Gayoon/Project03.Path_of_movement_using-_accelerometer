# 칼만필터 

```python
A = np.array([
    [1, dt, 0, 0],
    [0, 1, 0, 0],
    [0, 0, 1, dt],
    [0, 0, 0, 1],
])

H = np.array([
    [1, 0, 0, 0],
    [0, 0, 1, 0]
])
```

```python
# Q와 R은 튜닝 파라미터
# Q와 R을 변경하면서 칼만 필터의 성능을 개선 할 수 있음. 
Q = np.diag([1.0, 1.0, 1.0, 1.0])
print(Q)

R = np.array([
    [100000, 0],
    [0, 100000]
])
print(R)

P = 100 * np.eye(4)
print(P)
```

```python
x = np.array([0, 0, 0, 0])
```

```python
def KalmanTracking(x_, y_):
    global A, H, Q, R, P, x

    xp = A @ x
    Pp = A @ P @ transpose(A) + Q

    K = Pp @ transpose(H) @ inv(H @ Pp @ transpose(H) + R)

    z = (x_, y_)
    x = xp + K @ (z - H @ xp)
    P = Pp - K @ H @ Pp

    return x
```

### 수치적분

```python
def df_integral(data):  # 수치적분, 사다리꼴적분 가속도 -> 속도 -> 거리
    result = [0]

    for i in range(len(data) - 1):
        n = result[i] + ((data[i + 1] + data[i]) * (time[i + 1] - time[i]) / 2)  # 수치적분, 사다리꼴적분
        result.append(round(n, 7))
    return result


def df_rect(data):
    result = [0]

    for i in range(len(data) - 1):
        n = result[i] + (data[i + 1] * (time[i + 1] - time[i]))
        result.append(round(n, 7))
    return result


def df_simps(data):
    result = [0]

    for i in range(len(data) - 1):
        rect = result[i] + (data[i + 1] * (time[i + 1] - time[i]))
        integral = result[i] + ((data[i + 1] + data[i]) * (time[i + 1] - time[i]) / 2)

        n = rect * 2 / 3 + integral * 1 / 3
        result.append(round(n, 7))
    return result
```

# BPF : 센서 데이터에서 노이즈 제거

```python
# BPF
f1 = 0.5 / 10
f2 = 3 / 10
b, a = signal.butter(10, [f1, f2], btype='bandpass')
```

```python
dx = signal.lfilter(b, a, dx)
dy = signal.lfilter(b, a, dy)
```

```python
x_2 = df_simps(dx)  # x 가속도 -> 속도
x_3 = df_simps(x_2)  # x 속도 -> 거리
#
y_2 = df_simps(dy)  # y 가속도 - > 속도
y_3 = df_simps(y_2)  # y 속도 -> 거리
```
